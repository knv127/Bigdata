```python
from konlpy.tag import Komoran
komoran = Komoran()
text = '하지만 빅데이터의 문제점은 바로 사생활 침해와 보안 측면에 자리하고 있다.'
komoran.pos(text)
```

```python
komoran.nouns(text)
```

```python
from konlpy.tag import Hannanum
hannanum = Hannanum()
text = '하지만 빅데이터의 문제점은 바로 사생활 침해와 보안 측면에 자리하고 있다.'
hannanum.pos(text)
```
