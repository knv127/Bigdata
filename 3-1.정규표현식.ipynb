# 정규 표현식

```python
import re     #regular expression 정규표현식
```

```python
re.findall(r'apple','I have an apple')
```

```python
re.findall(r'apple','I have an apple pineapple')
```

```python
re.findall(r'an','I have an apple pineapple')
```

```python
re.findall(r'e','I have an apple pineapple')
```

```python
re.findall(r'\?','I have an apple pineapple?')
```

```python
re.findall(r'apple',"I have an apple but you don't have apples")
```

```python
re.findall(r't',"I have an apple but you don't have apples")
```

```python
re.findall(r'a.', 'I have an apple ha ') # . 은 하나의 문자
```

```python
re.findall(r'a.c', 'abc auc axxc')
```

```python
re.findall(r'a..c', 'abc auc axxc')
```

```python
re.findall(r'ab?c', 'ac abc auc axxc') # ? 앞의 문자가 0개 또는 1개
```

```python
re.findall(r'ab?', 'ac abc auc axxc')
```

```python
re.findall(r'a.?c', 'ac abc auc axxc')
```

```python
re.findall(r'ab*c', 'ac abc auc axxc') # * 앞의 문자가 0번 이상 출현 (a와 c 사이에 b가 있어도 되고 없어도 되고)
```

```python
re.findall(r'ax*c', 'ac abc auc axxc')
```

```python
re.findall(r'ab+c', 'ac abc auc axxc') # + 앞의 문자가 1번 이상 출현
```

```python
re.findall(r'ax+c', 'ac abc auc axxc')
```

```python
re.findall(r'1.3', '13 123 143 193 1993 1423 1ab3')
```

```python
re.findall(r'19+3', '13 123 143 193 1993 1423 1ab3')
```

```python
re.findall(r'a[bcd]c', 'ac aac abc axc acc') # [] 대괄호 안의 문자들 중 하나
```

```python
re.findall(r'a[a-z]c', 'ac aac abc axc acc')
```

```python
re.findall(r'a[012]c', 'ac a0c a1c a7c')
```

```python
re.findall(r'a[0-9]c', 'ac a0c a1c a7c')
```

```python
re.findall(r'a[a-z0-9]c', 'ac a0c a1c a7c abc')
```

```python
re.findall(r'a[02]c', 'ac a0c a1c a7c abc')
```

```python
re.findall(r'a[0-9]c', 'ac a0c a1c a7c abc') # a와 c 사이에 숫자 하나
```

```python
re.findall(r'a[0-9]+c', 'ac a0c a1c a7c abc a2323c a12c') # a와 c 사이에 숫자 여러개 가능하게 만드는 패턴
```

```python
re.findall(r'[\w]+', 'ac a0c a1c a7c')   # \w: a-zA-Z0-9_ 영문자, 숫자, _
```

```python
re.findall(r'[0-9]+', 'abc123xyz')
```

```python
re.findall(r'[a-z]+', 'abc123xyz')
```

```python
re.findall(r'[a-z]+[0-9]+[a-z]+', 'abc123xyz, def35yza, xxy5zz, 023afds23')
```

```python
re.findall(r'[a-z]+([0-9]+)[a-z]+', 'abc123xyz, def35yza, xxy5zz, 023afds23') # 소괄호 ()는 그룹을 만들 때 사용
```

```python
re.findall(r'([a-z]+)[0-9]+([a-z]+)', 'abc123xyz, def35yza, xxy5zz, 023afds23')
```
